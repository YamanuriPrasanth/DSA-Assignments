## Sum of digit of a number using recursion

Given a number, we need to find sum of its digits using recursion.

Examples: 

Input: 12345  
Output: 15  
Explanation: Sum of digits â†’ 1 + 2 + 3 + 4 + 5 = 15  

Input: 45632  
Output: 20 

```python
def SOD(num: int) -> int:
        if num<10:
            return num
        else:
            return num%10+SOD(num//10)

#Time Complexity: O(logn)
# Space Complexity:O(n) n is number of digits    
```

```python
SOD(12345)
```

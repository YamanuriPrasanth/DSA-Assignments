### Pow(x, n)

Implement pow(x, n), which calculates x raised to the power n (i.e., xn).

 

Example 1:

Input: x = 2.00000, n = 10  
Output: 1024.00000  

Example 2:

Input: x = 2.10000, n = 3  
Output: 9.26100  

Example 3:

Input: x = 2.00000, n = -2  
Output: 0.25000  
Explanation: 2-2 = 1/22 = 1/4 = 0.25  
 

Constraints:

-100.0 < x < 100.0  
-231 <= n <= 231-1  
n is an integer.  
Either x is not zero or n > 0.  
-104 <= xn <= 104  

```python
#Approach 1: Brute Force using Recursion
class Solution:
    def myPow(self, x: float, n: int) -> float:
        if n==1:
            return x
        elif n==0:
            return 1
        elif (n>1) or (n<1):
            negative_number_flag=0
            if n<1:
                negative_number_flag=1
                n=n*-1
            result=x*self.myPow(x,n-1)
            if n%2==0:
                if negative_number_flag==1:
                    return 1/result
                else:
                    return result
            else:
                if negative_number_flag==1:
                    return 1/result
                else:
                    return result

#Time Complexity: O(n)
# Space Complexity: O(n)         
```

```python
#Approach 2: Divide and conquer Approach
class Solution:
    def myPow(self, x: float, n: int) -> float:
        if n==1:
            return x
        elif n==0:
            return 1
        elif (n>1) or (n<1):
            negative_number_flag=0
            if n<1:
                negative_number_flag=1
                n=n*-1
            result=self.myPow(x,n//2)
            final_result=result*result
            if n%2==0:
                if negative_number_flag==1:
                    return 1/final_result
                else:
                    return final_result
            else:
                if negative_number_flag==1:
                    return 1/(x*final_result)
                else:
                    return x*final_result
        
#Time Complexity: O(logn)
#Space Complexity: O(logn)
```
